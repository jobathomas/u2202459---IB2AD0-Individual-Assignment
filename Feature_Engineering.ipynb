Below is a feature engineering excercise I performed, with reference to 2_01_data_and_feature_engineering_in_pandas.ipynb and 2_02_Joining_DataFrames.ipynb

```python
# importing pandas for dataframe creation

import pandas as pd

# creating dictionaries which contain lists

dataset1 = {"age":[12, 22, 45],"name":["Jennifer","James","Bethany"],}

dataset2 = {"country":["UK","Sweden","Germany"],"Race":["Hispanic","Caucasian","Black-African"],}

# creating dataframes from the above dictionaries
df1 = pd.DataFrame(dataset1)

df2 = pd.DataFrame(dataset2)

# joining both dataframes

joined_df = df1.join(df2)

print(joined_df)
```
